##Exercise 1

Investigate on the difference between is and == given the following variables.

```python
a = 2
b = 2.0
c = a
```

```python
# Using ==
print(a == b)  # True, as the values are equal (2 == 2.0)
print(a == c)  # True, as the values are equal (2 == 2)

# Using is
print(a is b)  # False, as they are different objects with different types
print(a is c)  # True, as 'c' is assigned the same object as 'a'
```

##Challenge 1

Write a program which repeatedly reads numbers until the user enters "done". Once "done" is entered, print out the total, count, and average of the numbers. If the user enters anything other than a number, detect their mistake using try and except and print an error message and skip to the next number.

```python
Enter a number: 4
Enter a number: 5
Enter a number: bad data
Invalid input
Enter a number: 7
Enter a number: done
16 3 5.333333333333333
```

```python
total = 0
count = 0

while True:
    user_input = input("Enter a number: ")

    if user_input.lower() == "done":
        break

    try:
        number = float(user_input)
        total += number
        count += 1
    except ValueError:
        print("Invalid input. Please enter a valid number.")

if count > 0:
    average = total / count
    print(f"Total: {total}, Count: {count}, Average: {average}")
else:
    print("No valid numbers entered.")
```

##Challenge 2

Write another program that prompts for a list of numbers as above and at the end prints out both the maximum and minimum of the numbers instead of the average.

```python
# Initialize variables for maximum and minimum
max_number = None
min_number = None

while True:
    user_input = input("Enter a number: ")

    if user_input.lower() == "done":
        break

    try:
        number = float(user_input)

        # Update maximum and minimum
        if max_number is None or number > max_number:
            max_number = number
        if min_number is None or number < min_number:
            min_number = number
    except ValueError:
        print("Invalid input. Please enter a valid number.")

if max_number is not None and min_number is not None:
    print(f"Maximum: {max_number}, Minimum: {min_number}")
else:
    print("No valid numbers entered.")
```
