Exercise 2

Write your salary computation to give an employee 1.5 times the hourly rate for hours worked above 40 hours. Example:

```python
Enter Hours: 45
Enter Rate: 10
Pay: 475.0
```

```python
hours_worked = float(input('Enter Hours: '))
hourly_rate = float(input('Enter Rate: '))

if hours_worked <= 40:
    pay = hours_worked * hourly_rate
else:
    regular_hours = 40
    overtime_hours = hours_worked - regular_hours
    overtime_rate = 1.5 * hourly_rate
    pay = (regular_hours * hourly_rate) + (overtime_hours * overtime_rate)

print('Pay:', pay)
```

Challenge

Write a program to prompt for a score between 0 and 100 and print a grade using the following table:

```python
Score           Grade
[ 97, 100]      1.0
[ 93,  97)        1.25
[ 89,  93)        1.5
[ 85,  89)        1.75
[ 81,  85)        2.0
[ 77,  81)        2.25
[ 74,  77)        2.5
[ 0,   74)        3.0
```

```python
score = float(input('Enter a score between 0 and 100: '))

if 97 <= score <= 100:
    grade = 1.0
elif 93 <= score < 97:
    grade = 1.25
elif 89 <= score < 93:
    grade = 1.5
elif 85 <= score < 89:
    grade = 1.75
elif 81 <= score < 85:
    grade = 2.0
elif 77 <= score < 81:
    grade = 2.25
elif 74 <= score < 77:
    grade = 2.5
elif 0 <= score < 74:
    grade = 3.0
else:
    print('Invalid score. Please enter a score between 0 and 100.')
    exit()

print('Grade:', grade)
```

Exercise 1

Rewrite your pay computation with time-and-a-half for overtime and create a function called compute pay which takes two parameters (hours and rate).

```python
Enter Hours: 45
Enter Rate: 10
Pay: 475.0
```

```python
def compute_pay(hours, rate):
    if hours <= 40:
        pay = hours * rate
    else:
        regular_hours = 40
        overtime_hours = hours - regular_hours
        overtime_rate = 1.5 * rate
        pay = (regular_hours * rate) + (overtime_hours * overtime_rate)
    return pay

hours_worked = float(input('Enter Hours: '))
hourly_rate = float(input('Enter Rate: '))

total_pay = compute_pay(hours_worked, hourly_rate)
print('Pay:', total_pay)
```

Challenge

Write a function called computegrade that takes a score as its parameter and returns a grade as a string using the following table:

```python
Score           Grade
[ 97, 100]      1.0
[ 93,  97)        1.25
[ 89,  93)        1.5
[ 85,  89)        1.75
[ 81,  85)        2.0
[ 77,  81)        2.25
[ 74,  77)        2.5
[ 0,   74)        3.0

Run the program repeatedly to test the various different values for input.
```

```python
def computegrade(score):
    if 97 <= score <= 100:
        grade = '1.0'
    elif 93 <= score < 97:
        grade = '1.25'
    elif 89 <= score < 93:
        grade = '1.5'
    elif 85 <= score < 89:
        grade = '1.75'
    elif 81 <= score < 85:
        grade = '2.0'
    elif 77 <= score < 81:
        grade = '2.25'
    elif 74 <= score < 77:
        grade = '2.5'
    elif 0 <= score < 74:
        grade = '3.0'
    else:
        grade = 'Invalid score. Please enter a score between 0 and 100.'
    return grade

score_input = float(input('Enter a score between 0 and 100: '))
result_grade = computegrade(score_input)
print('Grade:', result_grade)
```
